# Decision Making: Detection of digital random signals

## Detection Theory.

**[Additive] Gaussian channel model:**

$$
x(a_n,t)\longrightarrow \boxed{\textrm{zero-mean AWGN},\, \mathfrak{N}_{\eta}(0,\sigma_{\eta})}\longrightarrow y(t )=x(a_n,t)+\eta(t);\, \forall t\negthinspace\in\negthinspace T
$$\
The set $\{a_n:n\in N\negthinspace\subset\negthinspace \mathbb{Z}^{+}\}$ is the *Alphabet*, having a finite amount of elements. In practice, $N=2^{m},\, m\negthinspace\in\negthinspace\mathbb{Z}^{+}$

Alphabet generation rules: $\{p(a_n)\}$, where $p(a_n)$ -- symbol probability.

Let's assume a noiseless channel.

$$
\boxed{\textrm{TX: } x(a_n,t)}\longrightarrow \boxed{\textrm{Transmission channel}}\longrightarrow \boxed{\textrm{RX: } y(t)=x(a_n,t)}
$$
$$
\boxed{\textrm{ } \{p(a_n)\}}\longrightarrow \boxed{\textrm{CX}}\longrightarrow \boxed{\textrm{ } \{p(a_n)\}}
$$

In case of a Gaussian channel model, we have:
$$
\boxed{\textrm{TX } \{p(a_n)\} }\longrightarrow \boxed{\textrm{CX  }p(\eta)}\longrightarrow \boxed{\textrm{ RX } \{p(a_n)\cup p(\eta) \}\Rightarrow\{p(y|a_n)\}}
$$

Correct symbol detection:

\begin{equation*}
p(a_i|a_i)=\int\limits_{G_i\subset \mathfrak{G}} {p(y|a_i) dy} 
\end{equation*}


Incorrect symbol detection:

\begin{equation*}
p(a_j|a_i)=\int\limits_{G_j\subset \mathfrak{G}} {p(y|a_i) dy}, \, \forall j=0,\ldots,N-1,\, j\neq i 
\end{equation*}


\begin{equation*}
	 \int\limits_{G_i\subset \mathfrak{G}} {p(y|a_i) dy} +\sum_{\forall j \neq i}\int\limits_{G_j\subset \mathfrak{G}} {p(y|a_i) dy} =1, \forall i
\end{equation*}


That is, regardless of the assigned value, one symbol must be detected within a time length $T_S$.

Let's assume $N = 2,$ and $P(a_0)\neq (a_1)$, then, the following equality holds:

\begin{align*}
P_e &= P(a_0)P(a_1|a_0) +P(a_1)P(a_0|a_1)\\
&\\
&=P(a_0)\int\limits_{D_1} {p(y|a_0) dy} + P(a_1)\int\limits_{D_0} {p(y|a_1) dy}
\end{align*}


Provided the value set of $\{P(a_i)\}$, minimization of $P_e$ should be unfolded over two different spaces: $D_0$ and $D_1$.\\ 

\cita{How to get an optimization framework within a single space?}.\\

Since integration across all spaces fulfills the meaning of complete probability density, then, we have: 

\begin{equation*}
\int\limits_{D_1} {p(y|a_i) dy}+\int\limits_{D_0} {p(y|a_i) dy}=1,\, i=0,1
\end{equation*}


Therefore, the optimization function can be expressed over either decision space, $D_1$ or $D_0$:

\begin{align*}
P_e 	&=P(a_0)\int\limits_{D_1} {p(y|a_0) dy} + P(a_1)\int\limits_{D_0} {p(y|a_1) dy}\\
& = P(a_0)\int\limits_{{D_1}} {p(y|a_0) dy} + P(a_1)\left({1-\int\limits_{{D_1}} {p(y|a_1) dy}}\right)
\end{align*}

In case of \textit{balanced} channels, symbols are generated with the same probability: $P(a_0)=P(a_1)=1/2$. 

\begin{align*}
	P_e = \frac{1}{2}\left({\int\limits_{D_1} {p(y|a_0) dy} + \big({1-\int\limits_{D_1} {p(y|a_1) dy}}\big)}\right)
\end{align*}


Besides, binary \textit{symmetric} channels assume: $P(a_0|a_1)=P(a_1|a_0)$. 

\begin{align*}
P_e &= \int\limits_{D_1} {p(y|a_0) dy} \\
&= \int\limits_{D_0} {p(y|a_1) dy}
\end{align*}

## Bayes [risk] detection.

Generally speaking, the whole detection performance can be related to a cost of decision making, associated with each possible event $P(a_i|a_j)$, termed \textit{Hypothesis}:


1. We hypothesize that symbol $a_0$ has been sent, and really it has been. So, we fix the decision making cost to $c_{00}$
1. Symbol $a_0$ has been sent, but $a_1$ has been. So, we fix the decision making cost to $c_{10}$
1. Symbol $a_1$ has been sent, but $a_1$ has been. So, we fix the decision making cost to $c_{11}$
1. Symbol $a_1$ has been sent, but $a_0$ has been. So, we fix the decision making cost to $c_{01}$


Provided the sets $\{c_{ij}; i,j\in N\}$, $\{P(a_i|a_j);i,j\in N\}$, and $\{P(a_i);i \in N\}$, let's introduce a function that minimizes the following decision making functional, termed [*Bayes*] *risk*:

\begin{align*}
R({a})&=\mathbb{E}\{ r (a_i) \}= \sum\limits_{n = 1}^N {r(a_i)p(a_i)}\\
&=\sum\limits_{n = 1}^N \sum\limits_{n = 1}^N{c_{ij}p(a_i)p(a_j|a_i)}\\
&=\sum\limits_{n = 1}^N \sum\limits_{n = 1}^N{c_{ij}p(x_i)}\int_{D_j}{p(y|x_i)dy}\\
{\min_{D_j}\{R({a})\}}&= \sum\limits_{n = 1}^N \sum\limits_{n = 1}^N{c_{ij}p(x_i)}\, {\min_{D_j}\left\{{\int_{D_j}{p(y|x_i)dy}}\right\}}
\end{align*} 


### **Binary detection:**

\begin{align*} 
R({a})&=c_{00}p(a_0)p(a_0|a_0)+c_{10}p(a_0)p(a_1|a_0)+c_{11}p(a_1)p(a_1|a_1)+c_{01}p(a_1)p(a_0|a_1) \\
&\\
&\qquad p(a_j|a_i)=\int\limits_{G_j} {p(y|a_i) dy}, \\
R({a})&=c_{00}p(x_0)\int\limits_{{D_0}} {p(y|a_0) dy}+
c_{10}p(a_0)\int\limits_{{D_1}} {p(y|a_0) dy} +
c_{11}p(a_1)\int\limits_{{D_1}} {p(y|a_1) dy}+
c_{01}p(a_1)\int\limits_{{D_0}} {p(y|a_1) dy}\\
& \int\limits_{{D_1}} {p(y|x_i) dy}=1-\int\limits_{{D_0}} {p(y|x_i) dy},\, i=0,1\\
&\textrm{let's assume the symbol }a_0\textrm{ is sent}\\
R({a})&=p(a_0)c_{10}+p(a_0)c_{11}+\int\limits_{D_0}
{\left(p(a_1)\left(c_{01}-c_{11}\right)p(y|a_1)-\left(p(a_0)(c_{10}-c_{00}\right)p(y|a_0)\right)dy}\\
 {\min_{y}\{R({a})\}}&={\min_{y}} \left\{
 p(a_0)c_{10}+p(a_0)c_{11}+\int\limits_{D_0}
{\left(p(a_1)\left(c_{01}-c_{11}\right)p(y|a_1)-\left(p(a_0)(c_{10}-c_{00}\right)p(y|a_0)\right)dy}\right\}\\
&\qquad{\frac{d(p(a_0)c_{10}+p(a_0)c_{11})}{dy}}+\frac{\int\limits_{D_0}
	{\left(p(a_1)\left(c_{01}-c_{11}\right)p(y|a_1)-\left(p(a_0)(c_{10}-c_{00}\right)p(y|a_0)\right)dy}}{dy} = 0
\end{align*}


\begin{align*}
	{\min_{y}\{R({a})\}}:&\qquad
	 p(a_1)\left(c_{01}-c_{11}\right)p(y|a_1)-p(a_0)\left(c_{10}-c_{00}\right)p(y|a_0)<0\\ 
	 &\qquad\frac{p(y|a_1)}{p(y|a_0)}<\frac{p(a_0)(c_{10}-c_{00})}{p(a_1)(c_{01}-c_{11})}\\
	 &\\
 \textrm{since }R({a})\longrightarrow -\infty,&\qquad \qquad \textrm{successful: } c_{10}>0,\, c_{01}>0,\qquad \textrm{unsuccessful: }c_{00}\leq 0,\, c_{11}\leq 0\\
 &\\
 &\qquad\frac{{p(y|a_1 )}}{{p(y|a_0 )}}\substack{a_1\\>\\<\\a_0}
 \frac{p(a_0)(c_{10}-c_{00})}{p(a_1)(c_{01}-c_{11})}=\gamma \\
 &\\
 \varLambda (y) &{\buildrel \Delta \over =
 \frac{p(a_0)\left(c_{10}-c_{00}\right)}{p(a_1)\left(c_{01}-c_{11}\right)}}\qquad\textrm{Likelihood ratio}
 \end{align*}


Let's assume $c_{ij}-c_{ii}= \operatorname{const.};i\neq j$, 

\begin{align*}
	\varLambda (y)&= 1\Rightarrow \frac{{p(y|a_1 )}}{{p(y|a_0 )}}\substack{a_1\\>\\<\\a_0}\,1\\
	&\\
	a^{*}_i&={\mathop {\max}\limits_j \left\{ p(y|a_j) \right\} };\quad\textrm{Maximum a posteriori estimation}
\end{align*}

Let's model $ y= a_i + \eta$, $a_i = \operatorname{const}$, $i=\{0,1\}$ by a conditional pdf, $p(y|a_i)$, as a linear combination of two independent Gaussian variables:

$$
\mathfrak{N}({\mu_y,\sigma_y})= \mathfrak{N}({\mu_0,\sigma_0})+ \mathfrak{N}{(\mu_1,\sigma_1)},\quad w_i\in\mathbb{R}^{+}
$$

\\

**Combinacióon lineal de variables Gaussianas independientes:**

Sea el conjunto de variables $\{x_n;n\in N\}$ mutuamente independientes, cada una con distribución normal del tipo $\mathfrak{N}_{x_n}(\mu_{x_n},\sigma_{x_n})$. Entonces, su combinación lineal $ y=\sum\nolimits_{\forall n\in N}w_nx_n, $ con $w_n\in \mathbb{R},$ corresponde a otra variable Gaussiana.

Se introduce una forma alterna a la funci\'on densidad de probabilidad como descriptor de aleatoriedad atravez de la *función generadora de momentos*

$$
\mathrm{m}_x(\xi) = \mathbb{E} \{\exp(\xi x)\}
$$
que corresponde al valor esperado de la variable aleatoria proyectada $\exp(\xi x)$ y para la cual se cumple que:

\begin{align*}
\mathrm{m}_y(\xi) &=\mathbb{E}\{\exp(\xi (\sum_{\forall n\in N}w_nx_n))\}\\
&= \mathbb{E}\{\exp(\xi w_1x_1)\}\cdots\mathbb{E}\{\exp(\xi w_Nx_N)\}\\
& = \mathrm{m}_{x_1}(w_1\xi)\cdots\mathrm{m}_{x_N}(w_N\xi)\\
& = \prod_{\forall n\in N}\mathrm{m}_{x_n}(w_n\xi)
\end{align*}


Por cuanto, para cada variable normal se tiene que $\mathrm{m}_{x_n}(\xi) =\, \exp(\mu_{x_n}\xi +\sigma^{2}_{x_n}\xi^{2}/2)$, luego,

\begin{align*}
\prod_{\forall n\in N}\mathrm{m}_{x_n}(w_n\xi)& = \prod_{\forall n\in N}\exp(\mu_{x_n}\xi +\sigma^{2}_{x_n}\xi^{2}/2)\\
&= \exp\left (\xi(\sum_{\forall n\in N}w_n\mu_{x_n})+\xi^{2}(\sum_{\forall n\in N}w^{2}_n\sigma^{2}_{x_n})\right )\\
&= \exp(\mu_{y}\xi +\sigma^{2}_{y}\xi^{2}/2)
\end{align*}

con lo cual, la variable resultante $y$ tambi\'en es Gaussiana $\mathfrak{N}_{y}(\mu_{y},\sigma_{y})$ con momentos:

\begin{align*}
\mu_{y}&= \sum _{\forall n\in N}w_n\mu_{x_n}\,\\
\sigma^{2}_{y}&= \sum _{\forall n\in N}w^{2}_n\sigma^{2}_{x_n} 
\end{align*}


Make $w_1=1 - w_0, w_i\in[0,1]$: 

\begin{align*}
	p(y|a_i):\mathfrak{N}({\mu_y,\sigma_y})&= w_0\mathfrak{N}({\mu_0,\sigma_0})+ w_1\mathfrak{N}{(\mu_1,\sigma_1)}, 
%	\\	&=\mathfrak{N}({0,\sqrt{\sigma^{2}_0+\sigma^{2}_1}})
\end{align*} 


**Threshold computation:** $y_{\operatorname{opt}}=y\big |_{dR/dy = 0}$, that is, 

\begin{align*}
\frac{{p(\,y_{\operatorname{opt}}|a_1 )}}{{p(\,y_{\operatorname{opt}}|a_0 )}}&= \gamma\\
\frac{(\sqrt{2\pi} {\sigma}_1)^{-1}\exp(-(2 {\sigma}^{2}_1)^{-1}(y-\mu_1)^{2})}
{(\sqrt{2\pi} {\sigma}_0)^{-1}\exp(-(2 {\sigma}^{2}_0)^{-1}(y-\mu_0)^{2})}&=\gamma\\
\exp(-(2 {\sigma}^{2}_1)^{-1}(y-\mu_1)^{2} + (2 {\sigma}^{2}_0)^{-1}(y-\mu_0)^{2}) & =\frac{ {\sigma}_0}{ {\sigma}_1}\gamma\\
( {\sigma}^{2}_0)^{-1}(y-\mu_0)^{2} - ( {\sigma}^{2}_1)^{-1}(y-\mu_1)^{2} &= 2\ln ({ {\sigma}_0}\gamma/{ {\sigma}_1})\\
 \boxed{{\sigma}^{2}_1(y-\mu_0)^{2} - {\sigma}^{2}_0(y-\mu_1)^{2}}& = \boxed{2 {\sigma^{2}}_1 {\sigma^{2}}_0\ln ({ {\sigma}_0}\gamma/{ {\sigma}_1})}
\end{align*}

Make $ {\sigma}_0= {\sigma}_1 = {\sigma}.$ Compute $y_{\operatorname{opt}}$, assuming i) $\mu_1\neq -\mu_0$, ii) $\mu_1= -\mu_0$.

\begin{align*}
{\sigma}^{2} (y-\mu_0)^{2} - {\sigma}^{2} (y-\mu_1)^{2}& = 2 {\sigma^{2}} {\sigma^{2}}\ln ({ {\sigma} }\gamma/{ {\sigma}})\\
(y-\mu_0)^{2} - (y-\mu_1)^{2}& = 2 {\sigma^{2}}\ln ( \gamma )\\
2y(\mu_1-\mu_0)&={\mu_1}^{2}-{\mu_0}^{2}+2{\sigma^{2}}\ln ( \gamma )\\
y_{\operatorname{opt}}&=\frac{{\mu_1}^{2}-{\mu_0}^{2}+2{\sigma^{2}}\ln ( \gamma )}{2(\mu_1-\mu_0)}
\end{align*}


Make $P(a_0)= P(a_1)$, and $c_{ij}-c_{ii}= \operatorname{const.};i\neq j$. So, $\gamma=1$, and $\ln ( \gamma )=0,$
$$
y_{\operatorname{opt}} = (\mu_{0} + \mu_{1})/2
$$

Replacing $\mu_1= -\mu_0$, we have:
$$
{y_{\operatorname{opt}} = 0}
$$

## Optimal detection of binary signals

\begin{align*}
\boxed{\textrm{TX: \quad} \{p(a_n) \}: x( { {a}_n},t) }&\longrightarrow \boxed{\textrm{CX: \quad }y(t)=x( { {a}_n},t) +\eta(t):\mathfrak{N}_{\eta}(0,\sigma_{\eta}) }\longrightarrow \boxed{\textrm{RX: \quad} \textrm{detector: } \hat{a}_n=a_m\Rightarrow\{p(y|a_m)\}}\\
&\\
y(t) &= x(a_n,t) + \eta(t),\quad t\in T,\; n\in N=2^{m},\,m\in \mathbb{Z}^{+}
	\end{align*}

Main assumptions:
* The signal is perturbed by AWGN, $\mathfrak{N}_{\eta}(0,\sigma_{\eta})$, with Power Spectral Density $N_0$.
* PDF of generated symbols $\{P(a_n)\}$ is known. 
* At any time, one and just one symbol is transmitted through the channel with finite bandwidth $\Delta F$.

### Maximum likelihood principle: 

Computation of $p(y|a_n)$. Since $y(t) = x(a_n,t) + \eta(t)$, then

\begin{align*}
\eta(t) &= y(t)-x(a_n,t),\qquad\textrm{by denoting } {\xi(t)\triangleq \xi_k} \\
{\eta_k}&= {y_k-x_k(a_n)= \Delta y_k(n)}
\end{align*}


In other words,

## Detection of a signal with *unknown* constant:

\begin{align*}	
	H_0:\quad &y_k = \eta_k,\quad \textrm{}\\
	H_1:\quad &y_k= \alpha x_k + \eta_k; \quad\forall k\in[0,K-1]  
\end{align*}

where $x_i$ is a given function,  $\alpha$ is unknown constant, and $\eta_k$ is White Gaussian Noise having a known variance $\sigma^{2}_k$.
 
\begin{align*}
	 \Lambda (y|\hat{\alpha}) &=
	 \frac{P(y|\hat{\alpha})}{P(y)}= \frac{P({y}|\hat{\alpha})}{P({y})}
	\quad> \quad\gamma
\end{align*}

where the a posterior probabilities are modeled as:

\begin{align*}
	P({y}|\hat{\alpha})&=\frac{1}{(2\pi \sigma^{2}_{\eta})^{K/2}}\exp\left(-\frac{1}{2\sigma^{2}_{\eta}}\big({y}-\alpha{x}\big)^{\top}\big({y}-\alpha{x}\big)\right)\\
	P({y})&=\frac{1}{(2\pi \sigma^{2}_{\eta})^{K/2}}\exp\left(-\frac{1}{2\sigma^{2}_{\eta}} {y}  ^{\top} {y}  \right)\\
	{y}=[y_0,y_1,\ldots,y_{K-1}]^{\top},&\quad 	{x}=[x_0,x_1,\ldots,x_{K-1}]^{\top}
\end{align*}

and $\hat{\alpha}$ is the value of $\alpha$, estimated by the Maximum Likelihood  (ML) rule as follows:

\begin{align*}
	\frac{\partial }{\partial \alpha}\ln P({y}|\hat{\alpha})&=0 	\\
	\frac{\partial }{\partial \alpha}({y}^{2}-2\alpha{x}{y}+\alpha^{2}{y}^{2})&=0\\	
	-2 {x}{y}+\alpha {y}^{2}&=0
\end{align*}

So, the optimal value yields the estimation $\hat{\alpha}$:

\begin{align*}
 	{\hat{\alpha}}&   ={\frac{{y}^{\top}{x}}{{x}^{\top}{x}}}=  \frac{1}{\|{x}\|}{{y}^{\top}{x}}=	
 	\boxed{\sum_{\forall k\in K}{y_kx_k}/\sum_{\forall k\in K}{x^{2}_k}}
\end{align*}


Now, we calculate the log of ML 
inequality, $\Lambda ({y}|\hat{\alpha}) >\gamma$:

\begin{align*}
-\frac{1}{2\sigma^{2}_{\eta}}\sum_{\forall k\in K}{(-2\hat{\alpha}x_ky_k +\hat{\alpha}^{2}{x}^{2})}&>\ln \gamma
\end{align*}

Replacing $ {\alpha}$ by its estimate  of $\hat{\alpha}$:

\begin{align*}
	-\frac{1}{2\sigma^{2}_{\eta}}\left(-2\hat{\alpha}\hat{\alpha}\sum_{\forall k\in K}{x^{2}_k  +\hat{\alpha}^{2}}\sum_{\forall k\in K}{x^{2}_k}\right)&>\ln\gamma
\end{align*}

Therefore, the hypothesis $H_1$ about the  present  signal holds:

\begin{align*}
	\hat{\alpha}^{2}\quad&\substack{H_1\\>\\<\\H_0}\quad\frac{2\sigma^{2}_{\eta}\ln\gamma}{\sum_{\forall k\in K}{x^{2}_k}}\\
	\sum_{\forall k\in K}{x^{2}_ky^{2}_k}\quad& \substack{H_1\\>\\<\\H_0}\quad 2\sigma^{2}_{\eta}\ln\gamma\\
	\boxed{\big |\sum_{\forall k\in K}{x _ky _k}\big |}\quad& \substack{H_1\\>\\<\\H_0}\quad \boxed{\sqrt{2\sigma^{2}_{\eta}\ln\gamma}}
\end{align*}

Issue in implementing the previous decision-making algorithm: vector ${x}$ is not available practically. 

_Practical solution_:

\begin{align*}
\mathfrak{N}(\mu_1,\sigma_{\eta})+\mathfrak{N}(\mu_2,\sigma_{\eta}) &= \mathfrak{N}((\mu_1+\mu_2)/2,\sigma_{\eta})\\
\Rightarrow\mathfrak{N}(0,\sigma_{\eta})+\mathfrak{N}(\mu_2,\sigma_{\eta}) &= \mathfrak{N}(\mu_2/2,\sigma_{\eta})
\end{align*}



Thus, making $K_{\alpha}\ge K$, we estimate the amplitude as follows: 	
$$
{\hat{\alpha}}\simeq 2 \sum_{\forall k\in K_{\alpha}}{y_k}
$$

## Information Theory: Fundamentals

Probability as a uncertainty measure: Half of the interval is ascendant, while another half is descendant, resulting in two inputs with the same uncertainty value. Let's assume $\mathfrak{I}(x)\sim p(x)$

$$
p(x_0), x_0\in x_{min},x_{max}] \ne p'(x_0),\,x_0\in x'_{min},x'_{max}]

Figure 7: On the relative behavior of probabilistic measures

That is, having the same probability distribution, the uncertainty value of $x_0\in x$ is relative since it depends on the interval, within the variable $x$ is spanned.\\

May we have another uncertainty measure with asymptotically unchangeable  behavior?

**Premises for a new uncertainty measure:**


* It must reflect the concept of frequency, that is, it should be derived from the probability's axioms: \[\mathfrak{I}(x)\sim f (p(x))\] 
* It must be positive semidefinite, 
	 $$\mathfrak{I}(x)\in \mathbb{R}^{+}$$
	 where $\mathfrak{I}(x)=0$, whenever there is no uncertainty, $p(1),p(0)$
* It must asymptotically unchangeable,
	 $$\mathfrak{I}(x_i)>\mathfrak{I}(x_i), \quad \forall p(x_i)<p(x_j)$$
* It must be accumulative (positively): 
$$\mathfrak{I}(x_i)+\mathfrak{I}(x_i)\le\mathfrak{I}(x_i),\quad \mathfrak{I}(x_i)+\mathfrak{I}(x_i)\le\mathfrak{I}(x_ij)$$
