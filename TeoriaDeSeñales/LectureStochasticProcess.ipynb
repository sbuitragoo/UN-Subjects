## Random functions [processes]. 
A variable $x(\cdot,\cdot,)$ spanned over a domain(s) like time, space, etc. $x_i(t)$ is a trajectory. $\{x_i(t:\forall i\in N)\}$ is the signal ensemble.

Generally speaking, $p(x,t)=\operatorname{var}$, that is the uncertainty model varies over time. So, the random process is termed *nonstationary*. 

\begin{align*}
\textrm{Narrow-sense Stationarity: }p\left(\xi_i ,t\right) &= p\left(\xi_i ,t + \Delta t\right),\; \forall \Delta t\\
\textrm{Wide-sense Stationarity: }m_{1x}(t)&= m_{1x},\quad \sigma_{x}(t)= \sigma_{x}, \forall t
\end{align*} \\

**Wide-sense stationarity**. As a rule, $p(x)$ is not available. 

\begin{align*}
{\displaystyle\int\limits_{-\infty}^{\infty}}
x^{n}p_{\xi}\left( x\right) d x &\thickapprox {\lim\limits_{T\rightarrow \infty}\frac{1}{T}{\displaystyle\int\limits_{T}}
\xi^{n}\left(t\right)dt}\triangleq \overline{\xi^{n}\left(t\right)},\quad n\in {\mathbb{N}}: \textrm{Initial moment estimates} \\
{\displaystyle\int\limits_{-\infty}^{\infty}} \left(
x-m_{1\xi}\right) ^{n}p_{\xi}\left( x\right) d x &\thickapprox {\lim\limits_{T\rightarrow \infty}\frac{1}{T}{\displaystyle\int\limits_{T}} \left(
\xi\left(t\right)-\overline{\xi\left(t\right)}\right) ^{n}dt,}\quad n\geq2,\,n\in{\mathbb{N}}: \textrm{Centralized moment estimates} \\
{\displaystyle\int\limits_{-\infty}^{\infty}}{\displaystyle\int\limits_{-\infty}^{\infty}} 
x_{1}x_{2}p_{\xi_1 ,\xi_2 } \left( {x_1 ,x_2 }\right)d{ x_{1}}d{x_{2}} 
&\thickapprox {\lim\limits_{T\rightarrow \infty}\frac{1}{T}{\displaystyle\int\limits_{T}} {x_1 \left(t \right) } {x_2^{*}\left(t +\tau\right) } dt}= R_{x_1x_2} \left(\tau \right): \textrm{(Cross)Correlation function}\\
R_{\xi}\left( \tau\right) &
=\lim\limits_{T\longrightarrow\infty}\frac{1}{T}{\displaystyle\int\limits_{T}} 
\xi\left( t\right) \xi^{\ast }\left( t+\tau\right) 
dt : \textrm{ (Auto)Correlation function}\\
K_{\xi}\left( \tau\right) &
=\lim\limits_{T\longrightarrow\infty}\frac{1}{T}{\displaystyle\int\limits_{T}} \left( \xi\left(
t\right) -\overline{\xi\left(t\right)}\right) \left( \xi^{\ast }\left( t+\tau\right)
-\overline{\xi\left(t\right)}\right) dt: \textrm{ Covariance function}\\ &= R_{\xi}\left(
\tau\right)-m_{1\xi}^{2} 
\end{align*}

**Correlation function: Properties**

1. *Parity*.
	$
	R_{\xi}\left( \tau\right) =R_{\xi}^{*}\left( -\tau\right) ,\; K_{\xi}\left( \tau \right)= K_{\xi}(-\tau) 
	$
1. _Maximal value_.
	$ \left\vert R_{\xi}\left( \tau\right) \right\vert \leq R_{\xi}\left( 0\right) ,\quad \left\vert
	K_{\xi}\left( \tau\right) \right\vert \leq K_{\xi}\left( 0\right)$.

\begin{align*}
		K_{\xi}\left( 0\right) &= \frac{1}{T}{\displaystyle\int\limits_{0}^{T}} \left( \xi\left( t\right)
		-\overline{\xi\left(t\right)}\right) \left( \xi^{\ast }\left( t\right) -\overline{\xi\left(t\right)}\right)
		dt = \sigma_{\xi}^{2}, \\
		R_{\xi}\left( 0\right) &= \frac{1}{T}{\displaystyle\int\limits_{0}^{T}} \xi\left( t\right)
		\xi^{\ast}\left( t\right) dt = \overline{\xi^{2}\left( t\right)} %, \; \forall \overline{\xi\left(t\right)}=0
\end{align*}

3. _Periodicity_. Whenever, it holds that $\xi(t)=\xi(t-T),\, \forall t\in T$, then
$R_{\xi}\left( \tau\right) =R_{\xi}\left( \tau-T\right) ,\forall\tau\in T$. 

1. _Convergence_. Whenever, it holds that $\xi(t)\ne\xi(t-T),\, \forall t\in T$ then
$$
\lim\limits_{\left\vert \tau\right\vert \longrightarrow\infty}R_{\xi}\left( \tau\right)
=\overline{\xi^{2}\left( t\right)},\quad \lim\limits_{\left\vert \tau\right\vert
	\longrightarrow\infty}K_{\xi}\left(
\tau\right) =0, $$

1. _Shape restriction_. To be an implementable process, its Fourier Transform must fulfill the following condition:
\begin{equation}
{ \mathscr{F}}\left\{ R_{\xi}\left( \tau\right) \right\} \geq0,\quad\forall
\omega 
\end{equation}

**Power Spectral Density.**

According to Parseval's Theorem, we have:
$$
\frac{1}{2\pi}\int_{\!R}{X (\omega)X ^{*}(\omega)}d\omega= \int_{\!R}x^{2}dt $$


In cases of random signals, analysis is carried out within a long enough time segment:

\begin{align*}
x_T \left(t\right) &= {\mathop{\rm rect}\nolimits} _T \left(t\right)x \left(t\right),\\ X_{T}(\omega)&=\mathscr{F}\{x(t)\operatorname{rect}(t/T)\}\\&\\
\lim_{T\rightarrow\infty}\frac{1}{T}E\left\{\int_{T}x^{2}dt\right\}&=\lim_{T\rightarrow\infty}\frac{1}{2\pi}\int_{\!R}\frac{1}{T}E\left\{{X_{T} (\omega)X_{T} ^{*}(\omega)}\right\}d\omega\\
\textrm{assuming } m_{1x}=0, \quad m_{2x}&= \frac{1}{2\pi}\int_{\!R}S_{x}(\omega)d \omega,\\
{S_{x}(\omega)}&={\lim_{T\rightarrow\infty}\frac{|X_{T}(\omega)X_{T}^{*}(-\omega)|}{T}\geq 0,}\textrm{\quad PSD}\\
&\\
S_{x}(\omega)\in \!R^{+},&\quad S_{x}(\omega)= S_{x}(-\omega)
\end{align*}

**White Gaussian Noise.**

An ergodic process that holds all spectral components, each one with the same power in average is called  WGN, defined as follows:
\begin{equation*}
S\left(
\omega\right) =N_{0},\quad\omega\in\left( -\infty,\infty\right).
\end{equation*}

Using the Wiener-Jinchin Transform, we obtain: \\
\begin{equation*}
R\left( \tau\right) ={\displaystyle\int\limits_{-\infty}^{\infty}} \frac{N_{0}}{2}e^{j2\pi
	ft}df=\frac{N_{0}}{2}\delta\left( \tau\right) 
\end{equation*}

Assuming $\Delta\omega<\infty$, then, the following model is termed colored noise:
\begin{equation*}%\label{PDS:RuiRos}
\begin{cases}
S\left( \omega\right) &= N_{0},\left( -\Delta\omega<\omega<\Delta\omega\right) , \\
R\left( \tau\right) &=N_{0}\Delta\omega\operatorname{sinc}\left( 2\Delta\omega \tau\right) ,
\end{cases}
\end{equation*}
