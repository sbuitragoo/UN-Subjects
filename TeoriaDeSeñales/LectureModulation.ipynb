## Radio-Frequency Modulation RF

Baseband is translated on frequency domain by $\omega_c$:

\begin{align*}
y (t)= a (t) \Re\{\exp ( j\theta (t))\},
\end{align*}

$a(t)$ -- envelope, $\theta (t) = \omega_c t+\phi(t)$ -- instantaneous phase, $\phi(t)$ -- phase deviation, $\omega_c$ -- carrier frequency, $f_E:\mathbb{C} \mapsto \mathbb{C}$.

\begin{align*}
	a(t) &= f_E(x,t), \quad\phi(t)\sim \operatorname{const}, \quad \textrm{Envolvent (Amplitude) Modulation AM} \\
	\phi(t) &= f_E(x,t),\quad a(t)\sim \operatorname{const}, \quad \textrm{Angle Modulation}
	\end{align*} \\

**[Linear] Amplitude Modulation AM:** $\phi(t)$ is commonly assumed as a zero-valued term. AM is termed linear since the following linear dependency holds:
\begin{align*}
a(t)&\sim K x(t) + K_d;\,K,K_d=\operatorname{const}\\
\Rightarrow y(t)&= (K x(t) + K_d)\cos \omega_ct,\\
\Rightarrow Y(\omega) &= \mathscr{F}\{ {y (t)}\} = K\left(X (\omega+\omega_c) +
X(\omega-\omega_c)\right)/2 + K_d\left(\delta (\omega+\omega_c) +
\delta(\omega-\omega_c)\right)/2,
\end{align*} 

The first term contains the modulated information, while the second one does not.

In the last term, $k_d$ is chosen as to fulfill the unipolarity restriction of the envelope representation: 

\begin{equation}
 \operatorname{sgn}(a(t))=\operatorname{const},\forall t 
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2.8,14,1000)
#t = np.linspace(-2.8,4*np.pi,1000)
x = np.sin(t)/t

carrier = np.cos(5*2*np.pi*t)

modulated = x*carrier

def ed(w,a,b, q):
  e = np.zeros(len(w))
  if q == 1:
    for i in range(len(w)):
      if w[i] == 0:
        e[i] = a
      else:
        e[i] = 0
    return e
  elif q == 2:
    for i in range(len(w)):
      if w[i] == b or w[i] == -b:
        e[i] = a
      else:
        e[i] = 0
    return e

w = np.arange(-4,5,1)
esp = ed(w,1.5,1,1)
esp2 = ed(w,1.5,2,2)

w1 = np.linspace(-4,4,1001)
delta = ed(w1,2,1,1)

line1 = np.linspace(1,1.8,10)
line2 = np.linspace(2.3,3,10)

lt = -0.35*np.ones(10)

plt.figure(figsize=(16,9),dpi=90)
plt.subplot(2,2,1)
plt.plot(t,x,color='black')
plt.axhline(y=0.12, xmin=-3, xmax=15,linestyle=':',color='black',alpha=0.5)
plt.axis('off')
plt.title('(a) Modulating Signal $x(t)$',fontsize=14)
plt.subplot(2,2,2)
plt.plot(t,modulated,color='black')
plt.plot(t,x,color='black',linestyle='--')
plt.plot(t,-x,color='black',linestyle='--')
plt.axhline(y=0, xmin=-3, xmax=15,linestyle=':',color='black',alpha=0.5)
plt.axis('off')
plt.title('(b) Modulated Signal by $\cos(\omega_c t)$',fontsize=14)
plt.subplot(2,2,3)
plt.plot(w,esp,color='black')
plt.plot(w1,delta,color='black')
plt.text(0.9, -0.2, '$\omega$', fontdict=None, withdash=False)
plt.text(-1.3, -0.2, '$-\omega$', fontdict=None, withdash=False)
plt.title('(c) Energy Espectral Density of $x(t)$')
plt.axis('off')
plt.subplot(2,2,4)
plt.plot(w,esp2,color='black')
plt.plot(w1,delta,color='black')
plt.text(1.9, -0.2, '$\omega$', fontdict=None, withdash=False)
plt.text(-2.3, -0.2, '$-\omega$', fontdict=None, withdash=False)
plt.axvline(x=2,ymin=0,ymax=2,linestyle=':',color='black')
plt.axvline(x=-2,ymin=0,ymax=2,linestyle=':',color='black')
plt.axhline(y=1.5,xmin=-2,xmax=2,linestyle=':',color='black')
plt.axis('off')
plt.title('(d) Energy Espectral Density of Modulated Signal')
plt.arrow(0.5,0.5,1,0.025,width=0.01,length_includes_head=True,
          head_width=0.15,head_length=0.5,color='black')
plt.text(0.5, 0.65, 'Lower \nSide \nBand', fontdict=None, withdash=False)
plt.arrow(3.4,0.5,-1,0.015,width=0.01,length_includes_head=True,
          head_width=0.15,head_length=0.5,color='black')
plt.text(3, 0.65, 'Upper \nSide \nBand', fontdict=None, withdash=False)
plt.text(1.9, -0.4, '$2 \omega$', fontdict=None, withdash=False)
plt.plot(line1,lt,color='black')
plt.plot(line2,lt,color='black')
plt.scatter(1,-0.35,color='black')
plt.scatter(3,-0.35,color='black')
```

## Angle Modulation:

\begin{align*}
y(t) &= \Re\left\{a\exp(j\left(\omega_ct+\phi(t)\right))\right\}=
\Re\left\{a\exp(j\left(\omega_ct\right))\exp(j\phi(t))\right\}\\
y(t)&=\Re\left\{a\exp(j(\omega_c t + \phi(t)))\right\} = \Re\left\{a\exp(j\omega_ct) \left(\sum_{n=0}^{\infty}\frac{\left[(j\phi(t)\right)^n}{n!}\right)\right\}\\
& = \Re\left\{a\exp(j\omega_c t) \left(1 + j\phi(t)-{\phi^2(t)}/{2!}
- \cdots + {j^n\phi^n(t)}/{n!} + \ldots\right)\right\}.
\end{align*}

Broad-band Angle Modulation: find the bandwidth of Angle-modulated signals as the distance between $\omega_{\max}-\omega_{\min}$.\\
 Let $x(t)=a\cos\omega_mt$ be modulated under FM scheme. $\omega_m$ is the spectral margin marker.
 \begin{align*}
 \textrm{FM: } \omega_{i}(t)& = \omega_c + a k_f\cos\omega_m t = \omega_c +
 \Delta\omega\cos\omega_m t,\\
 \theta(t) &= \int_0^{t}\omega_i(\tau)d \tau = \omega_c t + \beta \sin
 \omega_m t,\\
 y_{FM}(t)&={\Re \left\{ Ae^{j\omega_ct}e^{j\beta\sin\omega_mt}
 \right\}}\Rightarrow\textrm{Broad-band FM-Modulated signal}
 \end{align*}
 Since $e^{j\beta\sin\omega_mt}$ is a periodic function, we can decompose it through the Fourier Series:
 \begin{equation*}
 e^{j\beta\sin\omega_mt}=\sum_{k=-\infty}^{\infty}F_ke^{jk\omega_mt},\quad\textrm{being}\quad F_k=\frac{1}{T}\int_{-T/2}^{T/2}e^{j\beta\sin\omega_mt}e^{-jk\omega_mt}d
 t,\, T = 2\pi/\omega_m
 \end{equation*}
 By noting $\xi=\omega_mt=(2\pi/T)t$, the following spectral values are obtained:
 \begin{equation*}
 F_k=\frac{1}{2\pi}\int_{-\pi}^{\pi}e^{j(\beta\sin\xi - k\xi)}d \xi.
 \end{equation*}
 Bessel function - canonical solution to differential equation ${\displaystyle x^{2}y''+xy'+(x^{2}-\nu ^{2})y=0,\,\nu \in \mathbb {C}}$
 For ${\displaystyle \nu \in \mathbb {R} }{\displaystyle \nu \in \mathbb {R} }$:

  1. $J_k(\beta)\in \!R$,
  1. $J_k(\beta)=J_{-k}(\beta)$, $\forall k$ even.
  1. $J_k(\beta)=-J_{-k}(\beta)$, $\forall k$ odd.
  1. ${\sum\nolimits_{k=-\infty}^{\infty}J^2_k(\beta)=1}$.

  \begin{align*}
e^{j\beta\sin\omega_mt}&=\sum_{k=-\infty}^{\infty}J_k(\beta)e^{jk\omega_mt},\\
y_{FM}(t)&=\Re \left\{ ae^{j\omega_ct}\sum_{k=-\infty}^{\infty}J_k(\beta)e^{jk\omega_mt} \right\}\\
&=a\sum_{k=-\infty}^{\infty}J_k(\beta)\cos\left((\omega_c+k\omega_m)t\right).
\end{align*}
