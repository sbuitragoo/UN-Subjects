# _Fourier Transform_ : Properties in signal analysis

\begin{equation*}
X(\omega)={\displaystyle\int\limits_{-\infty}^{\infty}} x(t)e^{-j\omega
	t}dt\overset{\vartriangle}{=}\mathscr{F}\left\{ x(t)\right\},\, x:\mathbb{R} \mapsto \mathbb{C}
\end{equation*}

Compute the _Energy Spectral Density_ (ESD) for a function $x(t)=a\operatorname{rect}_{\tau}(t)$, $a\in \!R, \tau\in\!R^{+}$.

\begin{equation*}
X(\omega) ={\int\limits_{-\infty}^{\infty}} a\operatorname{rect}_{\tau}(t)e^{-j\omega t} dt=a {\int\limits_{-\tau/2}^{\tau/2}}
e^{-j\omega t}dt =\dfrac{a}{j\omega}(e^{j\omega\tau/2}-e^{-j\omega\tau/2})=\dfrac{a\tau
	\sin(\omega\tau/2)}{\omega\tau/2}\\
 = {a\tau\operatorname{sinc}(\omega\tau/2)}
\end{equation*} \\

**_Linearity_**. Let $X_{n}(\omega)=\mathscr{F}$
$\{x_{n}(t)\},$ then, it holds that 
\begin{equation*}
	{\mathscr{F}\left\{ {\displaystyle\sum\limits_{n}} a_{n}x_{n}(t)\right\}
		={\displaystyle\sum\limits_{n}} a_{n}X_{n}(\omega),\quad
		\forall a_{n}=\operatorname{const.} }
	\end{equation*} \\

**_Symmetry_**. Let $x(t)=x_{\Re}(t)+jx_{\Im}(t)$, being $\mathscr{F}$ $\{x_{\Re}(t)\}=X_{\Re}(\omega)$ and $\mathscr{F}$ $\{x_{\Im
}(t)\}=X_{\Im}(\omega).$ \\

So, it holds that 
\begin{equation*}
{X(\omega)= X_{\Re}(\omega)+jX_{\Im}(\omega)}
\end{equation*} \\

**_Duality_**. Let $\mathscr{F}$ $\{x(t)\}=X(\omega),$ then 
	\begin{equation*}
  {\mathscr{F}\{X(t)\}=2\pi x(-\omega)}
  \end{equation*} \\

**_Scaling_**.  Let $\mathscr{F}\{x(t)\}=X(\omega),$ then 
\begin{equation*}
{\mathscr{F}\{x(\alpha t)\}= {X(\omega/\alpha)}/{\alpha},\,\forall \alpha\in\!R^{+}}
\end{equation*} \\

**_Time shifting (delay)_**.
\begin{equation*}{\mathscr{F}\{x(t-t_{0})\}=X(\omega)e^{-j\omega t_{0}},\, t_0\in\!R^{+}}
\end{equation*} \\


**_Frequency shifting (modulation or spectral translating)_**.
\begin{equation*}
{\mathscr{F}\{x(t)e^{\pm j\omega_{c}t}\}=X(\omega\mp\omega_{c}),\,\omega_{c}\in\!R^{+}} 
\end{equation*}

**_Convolution of two functions_**
\begin{align*}
{\mathscr{F}\left\{ x_{m}(t)x_{n}(t)\right\}} &={\left(
	X_{m}(\omega)\ast X_{n}(\omega)\right)/ {2\pi}}&\textrm{frequency domain}	 \\	
	{\mathscr{F}\left\{ x_{m}(t)*x_{n}(t)\right\}} &= {\left(
		X_{m}(\omega) X_{n}(\omega)\right)}&\textrm{time domain}
	\end{align*} \\

**_FT_: Remaining issues**.
* Lack of resolution because of inherent compromise between frequency and time resolution
* Low compactness of expansion
\end{itemize}

## Migrating from Fourier Mapping to Laplace Transform

> Let $x(t)\not\in \mathcal{L}_{2}(\!R)$ (a function not fulfilling Dirichlet conditions).

Values of signal $x(t)$ can be emphasized by their contribution differently (i.e., weighted) through the time. That is, $x(t)\simeq x(t)w(t)$, where $w(t)$ is the *weighting function*, for which $\|w(t)\|_2=1$. \\

Let's assume that the more close to \textit{x}-origin, the more the value contribution. In other words, very far values tend to be meaningless. 

Consider $w(t)= \operatorname{rect}_\tau(\beta t-\tau/2)\exp(-\alpha t)$, with $\alpha\in\!R^{+}$, and $\beta\mapsto \inf$. For simplicity, we focus on the quadrant I. \\
That is, a function that assigns a contribution attenuating as the values are more apart from the origin ("0"), but within a very wide interval.


For the assumed one-quadrant weighting function $u(t)e^{ - ct}$, with $c\in\!R^{+}$, ESD is rewritten as below:
\begin{equation}
X\left( { c;\omega} \right) = \int\nolimits_{ - \infty }^\infty {\left({x\left( t \right)u(t)e^{ - ct} } \right)e^{ - j\omega t} dt}
\end{equation}

Thus, the Fourier Transform is generalized by the Laplace Transform:

\begin{equation*}
\int\limits_0^\infty {x\left( t \right)e^{ - st} dt} = X\left( s \right) \triangleq \mathscr{L}\left\{ {x\left( t \right)} \right\},\quad s=c+j\omega 
\end{equation*} \\


> Therefore, LP does not require $x(t) \in \mathcal{L}^{2}(\!R)$ to be absolutely integrable on $\!R$.

Consider $w(t)= u(t)\exp(-\alpha t)$, with $\alpha\in\!R^{+}$, where $u(t)$ is the step function that allows weighting over the whole quadrant I. \\

FT can be though as LP evaluated on $j\omega$ (imaginary) axis, that is, $c=0$. \\

The FT projects (maps) a function of time $t$ into a complex valued function of real-valued domain $\omega$

The LP maps a function of time $t$ to a complex valued function of complex-valued domain $p$

## Short-time analysis of signals.

FT applies on a time segmented representation: $x_T(t)= x(t)\nu_T(t),\,\forall t\in T$. Windowing-based approaches deal with weighting functions, which have support compact\footnote{ function has non-zero values in a certain range and zeros elsewhere.} $T\in \!R^{+}$ (termed *window*).

The windowed Fourier transform (STFT) is obtained by by introducing a sliding window $\nu_T(t-u)$ that is designed to extract a small portion of the signal within $T$, as follows:
\begin{equation*}
	X_\nu(\omega)={\displaystyle\int\limits_{-\infty}^{\infty}} x(t)\nu_T(t-u)e^{-j\omega 
		t}dt
	\end{equation*}

A general symmetric apodization function $\nu(t)$ can be written as a Fourier series (cosine-sum window):
\begin{align*}
\nu_W(t)&= \nu_0+2\sum_{n=1}^\infty{\nu_n \cos(n \pi t/W)},\, W\in\!R^{+}\\
&\textrm{s.t.: } \nu_0+2\sum_{n=1}^\infty{\nu_n }=1
\end{align*}

when $\nu_0= 0.54,\nu_1=0.23$, Hamming window. (See other windows: Blackman, Nuttall, Gaussian, Poisson, Tukey, Dolph–Chebyshev).

## Wavelet Analysis.

Basis construction by scaling/translating principle: Generally, a Fourier series contains a single scale-varying basis set that is built by chopping up the same function using different scale sizes and translating values, as follows:

\begin{equation*}
x(t)= \sum_{n=0}^{\infty}{c_n\cos(n\omega_0 t-\theta_n)}, c_n\in \mathbb{C}
\end{equation*}

where $n\omega_0$ is the scaling factor and $\theta_n(t)$ is the translating term. 

Let $\psi\in \mathcal{L}_2(\!R)$ (termed \textit{mother wavelet}) a normalized zero-mean with support compact, for which using the dilation ($s\in\!R^{+}$) and translation ($u\in\!R$) properties a basis set is formed, and denoted by: 

\begin{align*}
	\bigg\{\psi_{s,u}(t)&= \frac{1}{\sqrt{s}}\psi(\frac{t-u}{s}) \bigg\}\\
	&\\
\textrm{Continuous Wavelet Transform: }\qquad& {\mathscr{F}\left\{ x(s,u)\right\}=\big<x(t),\psi^{*}(\frac{t-u}{s})\big> = \int_{\!R}{x(t)\frac{1}{\sqrt{s}}\psi^{*}(\frac{t-u}{s})} dt}
\end{align*} \\

Provided a basis set $\{\phi_n(t)\}$, there exists another set of basis $\{\hat{\phi}_n(t)\}$ (or \textit{dual function}) such that the following condition of orthonormality holds:
 \begin{equation*}
 	<\phi_n(t),\hat{\phi}_n(t)>={\displaystyle\int\nolimits_{\!R} {\phi_n(t)\hat{\phi}_m(t)}dt}= \delta_{mn}
 	\end{equation*}
Relying upon the orthonormal property, the spectral coefficients can be calculated by:
\begin{equation*}
	c_n=<x(t),\hat{\phi}_n(t)>
\end{equation*}

Mexican hat wavelet} with a complicated form that is derived from $\mathcal{N}(,)\partial\mathcal{N}(,)$ as follows:

\begin{equation*}
	\phi(t)=\frac{2}{\pi^{1/4}\sqrt{3\sigma}}{\big(\frac{t^{2}}{\sigma^{2}}-1\big)}\exp(-\frac{t^{2}}{\sigma^{2}})
	\end{equation*}

The wavelet function is designed to strike a balance between time domain (finite length) and frequency domain (finite bandwidth). As we dilate and translate the mother wavelet, very low frequency components are at large $s$ while very high frequency component are located precisely at small $s$.
