# Random: Probabilistic Models

## Random variables

Probabilistic space ${\left\{\varOmega,\mathcal{B},{P}\right\}}$: $\varOmega$ space of elementary events, $\sigma$-algebra $\mathcal{B}$ of Borell subsets from $\Omega$; frequency of occurrence $P\left(A\right)$, measured for $A\in \mathcal{B}$, having a unit-radius mass, $P\left(\varOmega\right)=1$. Kolmogorov axioms: \\

Lemma 1:
For ${\displaystyle A\in {\mathcal {F}}},\, {\displaystyle P(A)+P(A^{c})=1}.$ \\
Lemma 2:
For ${\displaystyle A_{1},\ldots ,A_{n}\in {\mathcal {F}}},\,
{\displaystyle P\left(\sum _{j=1}^{n}A_{j}\right)=\sum _{j=1}^{n}P(A_{j})}$. \\
Lemma 3:
For ${\displaystyle A,B\in {\mathcal {F}}},{\displaystyle P(A\cup B)=P(A)+P(B)-P(AB)}$ \\
Let the hypothesis set
$\left\{B_k:k=1,\ldots,n\right\}$ be a full group. So, it holds that (Bayes theorem):

\begin{align*}
P\left(B_k|A\right)&=\dfrac{P\left(B_k\right)P\left(A|B_k\right)}{P\left(A\right)}\\
&=\dfrac{P(B_k)P\left(A|B_k\right)}{\mathop \sum \limits_{k = 1}^n
	P\left(B_k\right)P\left(A|B_k\right)}
\end{align*}

$P\left(B_k\right)$ are marginal probabilities, $P\left(A|B_k\right)$ -- a prior probabilities, $P\left(B_k|A\right)$ -- posterior probabilities. 


**Characteristic functions**.
*Probability distribution* $ F_\xi \left( x \right) =
P_\xi \left(\xi < x:\xi \in \mathcal{B}\right) $. \\
_Probability function density_ describes the likelihood of occurring every possible value $x\in \xi$, which a random variable can assume $\xi$:

$$
P\left( {\xi = x} \right) = F_\xi \left( {x + \Delta x} \right) - F_\xi \left( x \right) > 0,\quad
\Delta x \to 0,
$$

\begin{equation*}%\label{RepEst:FunDistribucion}
\begin{cases}
F_\xi \left( x \right) &= {\displaystyle\int\limits_{ -\infty }^{x}} {p_\xi \left( \lambda \right)d\lambda },\\
p_\xi \left( \lambda \right)&=d F_\xi(\lambda)/d\lambda
\end{cases}
\end{equation*}

**Probability function density. Properties:**

1. $p_{\xi}\left( x\right) \geq0,\forall x\in \xi$

1. ${\displaystyle\int\limits_{\xi}} p_{\xi}\left( x\right) dx
	=F_{\xi}\left(\max\left\{x\right\}\right) = 1$,
1. ${\displaystyle\int\limits_{x_{1}}^{x_{2}}} p_{\xi}\left( x\right) dx
	=F_{\xi}(x_{2})-F_{\xi}(x_{1})=P\{x_{1}\leq \xi<x _{2}\}$

**Point estimate statistics.** Provided a measured set (sample) $\{x_k:k\in K\}$, compute in an approximate way (that is, estimate) a single characteristic value of random variables: 
$$
\tilde{\theta}=g_\theta(\{x_k:k\in K\},\epsilon)
$$

where $\tilde{\theta}$ is the best guess, and $\epsilon$ is the criterion of proximity (error). As a rule, 

$$
\epsilon^{2}=E\{(\tilde{\theta}-\theta)^{2}\}
$$

$g_\theta(\cdot)$ is the estimator. 

Two assumptions holds:

  *  pdf is unknown => non-parametric estimation 
  *  pdf is either assumed or imposed => parametric estimation
