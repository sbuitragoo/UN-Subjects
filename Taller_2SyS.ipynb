##2.Representación generalizada
* Encuentre el espectro $c_n \in \mathbb{C}$ de la señal $x(t)= \mu(t)$ a través de la representación generalizada $x(t)=\sum_{n=-\infty}^\infty c_n \phi_n(t)$, para $\phi_n(t)= e^{jn w_o t}$. Realice una simulación en python para $t \in [0,2]$ y determine el error para $n \in \{ 0, \pm1, \pm2 \}$

```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
import math as mt
import warnings
```

```
#Calcular Espectro
N = 50
nv = np.linspace(0, N, 2*N+1)
Cn = (((-1)**nv)-1)/(-1j*2*nv*np.pi)
```

```
#Graficar Espectro
fig = plt.figure(figsize=(4,8))
plt.subplot(3,1,1)
plt.stem(nv, np.real(Cn), 'r')
plt.title(r'Re{Cn}')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$Re\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(3,1,2)
plt.stem(nv, np.imag(Cn), 'r')
plt.title(r'Im{Cn}')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$Im\{C_n\}$',fontsize = 14)
plt.axis('tight')
plt.grid()

plt.subplot(3,1,3)
plt.stem(nv, abs(Cn), 'r')
plt.title(r'|Cn|')
plt.xlabel(r'$nw_o$[rad/s]',fontsize = 14)
plt.ylabel(r'$|C_n|$',fontsize = 14)
plt.axis('tight')
plt.grid()

fig.tight_layout()
```

```
#Error Relativo
Co2 = 0
Cn2 = 0
for n in range (-2,0):
  Cn = ((((-1)**n)-1)/(2*n*np.pi))*((((-1)**n)-1)/(2*n*np.pi)) 
  Cn2 = Cn2 + Cn
 
Px = 1/2
 
Er = 1 - (Co2 - (2*Cn2))/Px
print("El error es: ", Er)
```

###  4.1 Uso de la FFT para los ejercicios del punto 3.1

a) $x(t)=t$

```
T = 2*np.pi
Fs = 500
Nm = T*Fs
tv = np.linspace(-T/2,T/2,int(Nm))
xt = tv

vf = np.fft.fftfreq(len(xt),1/Fs) #calculo vector de frecuencias
Xw = np.fft.fft(xt) #cálculo FFT

plt.stem(vf,abs(Xw),'r')
plt.title(r'|X[n]|')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|$',fontsize = 14)
plt.axis('tight')
plt.grid()
plt.show()
plt.stem(vf,abs(Xw)/len(vf),'r')
plt.title(r'|X[n]|/K=c_n')
plt.xlabel(r'$f$[Hz]',fontsize = 14)
plt.ylabel(r'$|X[n]|/K$',fontsize = 14)
```
