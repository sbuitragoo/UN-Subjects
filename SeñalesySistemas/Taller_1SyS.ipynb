### 4. Discretización de señales cosenoidales.

* Se pretende muestrear la señal $x(t)= \cos(\Omega t)$ con $ t \in [0, T],\Omega = 2\pi F, F = 1/T, F = 60Hz$.  Se emplea un sistema de discretización con frecuencia de muestreo $F_s = 90Hz$. Demuestre si el sistema utilizado es apropiado para la señal $x(t)$ y estime la señal capturada. Realice la simulación en Python del proceso de discretización.

```
def g_cos(A,Fo,ti,tf,Fs):
    Ts = 1/Fs
    Fc = 20*Fs
    Tc = 1/Fc
    if tf > ti:
        vec_t = np.arange(ti,tf,Tc)
        #vec_t = np.linspace(ti,tf,(tf-ti)*Fc)
        xt = A*np.cos(2*np.pi*Fo*vec_t)
    else:
        assert(tf > ti), "tf > ti"

#%% discretizar
    vec_ts = np.arange(ti,tf,Ts)
    wo = 2*np.pi*Fo
    xs = A*np.cos(wo*vec_ts)
    plt.stem(vec_ts,xs,'b',label="x[nTs]")

#%% Graficar
    plt.figure(figsize=(6, 3))
    plt.plot(vec_t,xt,'r',linewidth=4,label="x(t)")
    plt.stem(vec_ts,xs,'b',label='x[nTs]')
    return xt, vec_t

#A = 1
#Fo = 60
#ti = 0
#tf = 3*(1/ Fo)
#Fs = 90
#xt,vec_t = g_cos(A,Fo,ti,tf,Fs)
xt,vec_t = g_cos(1,60,0,3*(1/Fo),90)
```

* Sea la señal Gaussiana $x(t)=e^{-at^{2}}$ con $a \in \mathbb{R^+}$, el sistema A con relación entrada-salida $y_A(t)=x^2(t)$, y el sistema lineal e invariante en el tiempo B con respuesta al impulso $h_B(t)=Be^{-bt^{2}}$: a) encuentre la salida del sistema en serie $x(t) \rightarrow h_B(t) \rightarrow y_A(t) \rightarrow y(t)$. b) Encuentre la salida del sistema en serire $x(t) \rightarrow y_A(t) \rightarrow h_B(t) \rightarrow y(t)$. 

Primero declaramos las funciones:

```
a, t, b, B, tau = sym.symbols("a t b B tau", real=True)
xf = sym.exp(-a*((t)**2))
hf = B*sym.exp(-b*((t)**2))
```

Hacemos el productoo de las dos funciones, reemplazamos las constantes por 1 y guardamos esto en una nueva variable.

```
s = xf.subs({t: tau,a: 1})*hf.subs({t: t-tau,B: 1, b: 1})
print(s)
```

Realizamos la convolución integrando el producto de las dos funciones respecto a $\tau$ entre $-\infty$,$\infty$ dejando una quieta y desplazando la otra.

```
conv = sym.integrate(s, (tau,-sym.oo, sym.oo))
print(conv)
```

Elevamos la convolución obtenida al cuadrado.

```
c2 = conv**2
print("La salida obtenida es y(t) =",c2)
```

Ahora gráficamos la salida del sistema:

```
yt = (mt.sqrt(2)*mt.sqrt(sym.pi)*(-sym.erfc(mt.sqrt(2)*t/2) + 2)*sym.exp(-t**2/2)/4 + mt.sqrt(2)*mt.sqrt(sym.pi)*sym.exp(-t**2/2)*sym.erfc(mt.sqrt(2)*t/2)/4)**2

sym.plot(yt, (t, -2, 2), ylim=[-0.2, 2.2])
```

Ahora para el punto b) elevamos $x(t)$ al cuadrado:

```
xfn = xf**2
xfn
```

Creamos una nueva variable con el producto entre $x(t)^2$ y $h_B(t)$

```
s1 = xfn.subs({t: tau,a: 1})*hf.subs({t: t-tau,B: 1, b: 1})
print(s1)
```

Procedemos a realizar la convolución entre $x(t)^2$ y $h_B(t)$:

```
conv2 = sym.integrate(s1, (tau,-sym.oo, sym.oo))
print("La salida obtenida es y(t) =",conv2)
```

Ahora gráficamos la salida del sistema:
